# Pyplot

Pyplot is a Matplotlib module which provides a MATLAB-like interface. Matplotlib is designed to be as usable as MATLAB, with the ability to use Python and the advantage of being free and open-source.

With the help of pyplot we can create so many plots like Line Plot, Histogram, Scatter, 3D Plot, Image, Contour, and Polar.

```python
import matplotlib.pyplot as plt 

x = [3,6,9,12]

y = [5,10,15,20]

plt.plot(x,y) 
plt.axis([0, 15, 0, 25]) 
plt.title('Example Graph')
plt.show() 
```

# Parameters: This function accepts parameters that enables us to set axes scales and format the graphs. These parameters are mentioned below :-

plot(x, y): plot x and y using default line style and color.


plot.axis([xmin, xmax, ymin, ymax]): scales the x-axis and y-axis from minimum to maximum values


plot.(x, y, color=’green’, marker=’o’, linestyle=’dashed’, linewidth=2, markersize=12): x and y co-ordinates are marked using circular markers of size 12 and green color line with — style of width 2


plot.xlabel(‘X-axis’): names x-axis


plot.ylabel(‘Y-axis’): names y-axis


plot(x, y, label = ‘Sample line ‘) plotted Sample Line will be displayed as a legend

```python
import matplotlib.pyplot as plt

year = [2013,2014,2015,2016,2017,2018,2019]

Rohit_runs = [450,520,415,650,750,800,900]

Virat_runs = [520,580,500,720,754,820,850]
  
plt.plot(year,Rohit_runs,'orange',label = 'Rohit',linewidth = 3.0)# change color and line width according to your requirement
plt.plot(year, Virat_runs,'b',label = 'virat',linewidth = 2.0) # change color and line width according to your requirement
  
# naming of x-axis and y-axis 
plt.xlabel('Years') 
plt.ylabel('Runs Scored')
  
# naming the title of the plot 
plt.title('Runs Scored by Virat and Rohit')
         
plt.legend() # Display the legend
plt.show() 
```

```python
#same graph with some line formatting

plt.plot(year,Rohit_runs,'orange',label = 'Rohit',linewidth = 2.0,marker ='o', markersize = 12)

plt.plot(year, Virat_runs,'g',label = 'virat',linewidth = 2.0,linestyle ='dashed',marker ='o', markersize = 12)

# For more info on markers visit https://matplotlib.org/api/markers_api.html

plt.xlabel('Years') 
plt.ylabel('Runs Scored')
  

plt.title('Runs Scored by Virat and Rohit')
         
plt.legend()
plt.show()
```

# Multiple figures

```python
import numpy as np

def create_plot(t): 
    # setting the x-axis vaues 
    return np.sin(2*np.pi*t) * np.cos(2*np.pi*t)

t = np.arange(0, 5, 0.01)

plt.figure()
# The subplot() command specifies numrows, numcols, 
# fignum where fignum ranges from 1 to numrows*numcols.
plt.sutbplo(211)
plt.grid()
plt.plot(t, create_plot(t), 'b-')

plt.subplot(212)
plt.plot(t, np.sin(2*np.pi*t), 'r--')
plt.show()
```

# Operations using Pandas

```python
# create a  random dataframe
data = {'city' : ['Mumbai','Delhi','Bengaluru','Hyderabad','chennai','kolkata','Jaipur','Pune'],
        'Temperature':[30,28,32,25,38,21,39,23],
        'Event':['Cloudy','Cloudy','Sunny','Rain','Sunny','Rain','Sunny','Rain']}
df = pd.DataFrame(data,columns = ['city','Temperature','Event'])
df
```

```python
x = df.groupby('Event')# grouping

#to display grouped items
for Event,Event_df in x:
    print(Event)
    print(Event_df)
```

```python
#go for a specific group

x.get_group('Cloudy') # play with it
```

# Concatenation and Merging datasets

```python
Ind_players = {'player':['Dhawan','Rohit','Virat','Shreyas','Dhoni'],
        'Score':[55,62,78,35,28]}
df = pd.DataFrame(Ind_players,columns = ['player','Score'])
df
```

```python
Aus_players = {'player':['Warner','Finch','Smith','Wade','Maxwell'],
              'Score':[42,35,80,21,18]}
df_1 = pd.DataFrame(Aus_players,columns = ['player','Score'])
df_1
```

```python
#concatenate two dataframes
new_data = pd.concat([df,df_1])
new_data #output will be a new dataset with repetative index
```

```python
# correcting index
new_data = pd.concat([df,df_1],ignore_index = True)
new_data
```

# Merge DataFrames

```python
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore", 'hyderabad'],
    "temperature": [32,45,30,40]})
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city": ["delhi","mumbai","banglore"],
    "humidity": [68, 65, 75]})
humidity_df
```

```python
#merge two dataframes with out explicitly mention index

df = pd.merge(temperature_df, humidity_df, on='city')
df
```

```python
#OUTER-JOIN
df = pd.merge(temperature_df, humidity_df, on='city', how='outer')
df
```
